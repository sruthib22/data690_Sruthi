```python
#loading the data
df = pd.read_csv('https://raw.githubusercontent.com/sruthib22/data690_Sruthi/main/Individual_project/wdi_data%20(4).csv')
df.head()
```

```python
# finding all the unique values of country code
df['Country Code'].unique()
```

```python
df['Country Name'].unique()
```

```python
# dropping the unwanted column
df_new= df.drop(columns=["Unnamed: 0"])
df_new.head()
```

```python
df_new.tail(20)
```

#### calculating the number of infant deaths in india 
  - This helps us evaluate the number of deaths of children and especially the number of girl child deaths compared to male child deaths due to gender descrimination in India

```python
df_ind = df_new[df_new['Country Code']== 'IND']
df_ind
```

```python
df_ind_infant = df_ind.query("indicator in(['SH.DTH.IMRT'])")
df_ind_infant
```

```python
df_pivot = df_ind_infant.pivot(index =['Year','Country Code','Country Name','Region','Income Group','Lending Type'], columns ='indicator', values ='value').reset_index()
df_pivot
```

```python
fig = px.bar(
    df_pivot.dropna(),
    x = "Year",
    y = "SH.DTH.IMRT",
    title = "Infant deaths",
    labels = {"SH.DTH.IMRT":"infant male deaths ","SH.DTH.IMRT":"total child deaths"},
    color = "Region",
    hover_name = "Country Name",
    text = "Country Name",
    #size = "SP.POP.TOTL"
)
#fig.update_traces(textposition = "top center")

fig.show()
```

#### comparing the number of infant deaths between different countries 
  - comparing the number of deaths between different countries.

```python
df_infants = df_new.query("indicator in(['SH.DTH.IMRT','SH.DTH.IMRT.FE','SH.DTH.IMRT.MA'])")
df_infants
```

```python
df_country = df_infants[df_infants["Country Code"].isin(["IND","CHN","PAK","USA"])]
df_country
```

```python
df_infants_pvt = df_country.pivot(index =['Year','Country Code','Country Name','Region','Income Group','Lending Type'], columns ='indicator', values ='value').reset_index()
df_infants_pvt
```

```python
fig = px.scatter_3d(data_frame= df_infants_pvt, x='SH.DTH.IMRT.FE', y='SH.DTH.IMRT.MA', z='SH.DTH.IMRT',
                    labels = {"SH.DTH.IMRT.FE":"infant female deaths ","SH.DTH.IMRT.MA":"infant male deaths", "SH.DTH.IMRT":"infant deaths" },
              color='Country Name', )
fig.show()
```

#### Retirement benifits
  -  to see whether both the genders are given equal benefits after retirement.

```python
df_retirement = df_new.query("indicator in(['SG.AGE.RTRE.PL.FE','SG.AGE.RTRE.PL.MA'])")
df_retirement
```

```python
df_pivot_ret = df_retirement.pivot(index =['Year','Country Code','Country Name','Region','Income Group','Lending Type'], columns ='indicator', values ='value').reset_index()
df_pivot_ret
```

```python
fig = px.pie(data_frame = df_pivot_ret, values='SG.AGE.RTRE.PL.FE', names='Country Name',title = "retirement benefits for women", hover_name= "Country Name" ,color_discrete_sequence=px.colors.sequential.RdBu)
fig.show()

fig = px.pie(data_frame = df_pivot_ret, values='SG.AGE.RTRE.PL.MA', names='Country Name',title = "retirement benefits for men", hover_name= "Country Name" ,color_discrete_sequence=px.colors.sequential.RdBu)
fig.show()
```
