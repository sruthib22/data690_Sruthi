```python
import pandas as pd
import plotly
print('plotly version',plotly.__version__)
import plotly.graph_objects as go
import plotly.express as px
import plotlywidget
```

```python
df = pd.read_csv('wdi_data.csv')
df.head()
```

```python
df_new= df.drop(columns=["Unnamed: 0"])
df_new.head()
```

```python
df_new.tail(20)
```

```python
df_ind = df_new[df_new['Country Code']== 'IND']
df_ind
```

```python
df_ind_infant = df_ind.query("indicator in(['SH.DTH.IMRT.MA','SH.DTH.IMRT.MA','SH.DTH.IMRT'])")
df_ind_infant
```

```python
df_pivot = df_ind_infant.pivot(index =['Year','Country Code','Country Name','Region','Income Group','Lending Type'], columns ='indicator', values ='value').reset_index()
df_pivot
```

```python
fig = px.bar(
    df_pivot.dropna(),
    x = "Year",
    y = "SH.DTH.IMRT.MA",
    title = "Infant deaths",
    labels = {"SH.DTH.IMRT.MA":"infant male deaths ","SH.DTH.IMRT":"total child deaths"},
    color = "Region",
    hover_name = "Country Name",
    text = "Country Name",
    #size = "SP.POP.TOTL"
)

#fig.update_traces(textposition = "top center")

fig.show()
```

```python
df_us = df_new[df_new['Country Code']== 'USA']
df_us
```

```python
df_us_edu = df_us.query("indicator in(['SE.PRM.CUAT.ZS'])")
df_us_edu
```

```python
df_pivot = df_us_edu.pivot(index =['Year','Country Code','Country Name','Region','Income Group','Lending Type'], columns ='indicator', values ='value').reset_index()
df_pivot
```

```python
fig = px.line(
    df_pivot.dropna(),
    x = "Year",
    y = "SE.PRM.CUAT.ZS",
    title = "Infant deaths",
    labels = {"SH.DTH.IMRT.MA":"infant male deaths","SE.PRM.CUAT.ZS":"educational attainment"},
    color = "Region",
    hover_name = "Country Name",
    text = "Country Name",
    #size = "SP.POP.TOTL"
)

fig.update_traces(textposition = "top center")

fig.show()
```
